```python
import pandas as pd

# 공공 데이터 포털에서 다운로드 받은 bike02.csv 파일을 읽어오는 방법
df = pd.read_csv('greenData/bike02.csv', encoding='cp949')
# 상위10개출력한다.
print(df.head(10))
```

```python
# 대여건수가1000 이상인데이터만추출
df[df['대여건수'] >= 1000]
```

```python
# 대여건수기준내림차순정렬
df.sort_values(by='대여건수', ascending = False)
```

```python
# 전체대여건수합계구하기
df.groupby('대여일자')['대여건수'].sum()
```
