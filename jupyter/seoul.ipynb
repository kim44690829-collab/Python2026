```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('greenData/seoul.csv', encoding='cp949')
df2 = pd.DataFrame(df)
# df2.head(10)
# df2.info()
# df2.describe()

df3 = df2.dropna()
yearFineDust = df3.groupby('측정년도')['미세먼지(㎍/㎥)'].mean().reset_index()
yearFineDust

sns.set_style('whitegrid')
sns.set_palette('pastel')


# -------------------------------------
# 반드시 한극 깨짐 현상 코드 삽입
# -------------------------------------

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize= (8,5))

plt.grid(linestyle = ':')

plt.bar(yearFineDust['측정년도'], yearFineDust['미세먼지(㎍/㎥)'], width=0.8, color='skyblue')

plt.title('측정년도별 미세먼지 평균')
plt.xlabel('측정년도')
plt.ylabel('미세먼지 평균값')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('greenData/seoul.csv', encoding='cp949')
df2 = pd.DataFrame(df)
# df2.head(10)
# df2.info()
# df2.describe()

df3 = df2.dropna()

yearUltraFineDust = df3.groupby('측정년도')['초미세먼지(㎍/㎥)'].mean().reset_index()
yearUltraFineDust

sns.set_style('whitegrid')
sns.set_palette('pastel')


# -------------------------------------
# 반드시 한극 깨짐 현상 코드 삽입
# -------------------------------------

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize= (8,5))

plt.grid(linestyle = ':')

plt.plot(yearUltraFineDust['측정년도'], yearUltraFineDust['초미세먼지(㎍/㎥)'], marker='o', markersize=4 ,color='skyblue', label='초미세먼지')

plt.title('측정년도별 초미세먼지 변화(seaborn)')
plt.xlabel('측정년도')
plt.ylabel('초미세먼지 평균값')

plt.legend()

plt.show()
```
