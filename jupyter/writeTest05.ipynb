병합(merge) : 최소 2개 이상의 데이터 프레임을 합치는 작업

```python
import pandas as pd

df1 = pd.DataFrame({
    "지역" : ["서울","부산","대구"],
    # "연도" : [2024, 2025, 2024],
    "총인구" : [1000000, 300000, 200000]
})
df2 = pd.DataFrame({
    "지역" : ["서울","부산","인천"],
    # "연도" : [2024, 2025, 2024],
    "병원수" : [150, 80, 60]
})

# 데이터 프레임의 필드명을 확인하는 함수 => .columns
print(df1.columns)
print(df2.columns)

# merge공식 => pd.merge(left_DataFrameName, right_DataFrameName, how="타입", on="기준 컬럼")
# how 방식 => inner (교집합), left(왼쪽 기준), right(오른쪽 기준), outer(합집합)
# merged = pd.merge(df1, df2, how="inner", on="지역") # => inner
# merged = pd.merge(df1, df2, how="outer", on="지역") # => outer
# merged = pd.merge(df1, df2, how="left", on="지역") # => left
merged = pd.merge(df1, df2, how="right", on="지역") # => right

# 컬럼명이 다를 경우 merge
# merged = pd.merge(df1, df2, how="inner", left_on = "지역", right_on = "지역명")

# 여러개의 컬럼명으로 병합하는 경우
# merged = pd.merge(df1, df2, how="inner", on = ['지역', '연도'])

print(merged)
```

결측치 : 데이터 병합하는 과정에서 데이터가 비어있는 상태를 의미한다.
즉, 값이 없어서 NaN(Not a Number)로 나타나는 경우를 말한다.

```python
import pandas as pd

df1 = pd.DataFrame({
    "지역" : ["서울","부산","대구"],
    "총인구" : [1000000, 300000, 200000]
})
df2 = pd.DataFrame({
    "지역" : ["서울","부산","인천"],
    "병원수" : [150, 80, 60]
})

# 병합
merged02 = pd.merge(df1, df2, how = "outer", left_on="지역", right_on = "지역")
print(merged02)
print()

# 결측치 개수 확인
# isna() => 결측치가 존재한다.
# print(merged02.isna().sum())

# 결측치 => 처리해줘야 그래프를 그릴 수 있음
# 1. 특정 값으로 채우기 : fillna(0), fillna(평균값)
# merged02['총인구'], merged02['병원수'] = merged02['총인구'].fillna(0), merged02['병원수'].fillna(0) => 가능하네
# merged02['총인구'] = merged02['총인구'].fillna(0)
# merged02['병원수'] = merged02['병원수'].fillna(0)
# print(merged02)

# 2. 행 삭제하기 => dropna() => 필요 없는 데이터일 경우에만
# merged02 = merged02.dropna()
# print(merged02)

# 3. 평균 / 중간값으로 채우기
merged02['총인구'] = merged02['총인구'].fillna(merged02['총인구'].mean())
merged02['병원수'] = merged02['병원수'].fillna(merged02['병원수'].mean())
merged02['비율'] = merged02['총인구'] / merged02['병원수']
merged02.style.format({'총인구' : '{:,.0f}', '병원수':'{:.0f}', '비율':'{:.1f}'})
```
