###### 과목별 점수를 세로 막대그래프로 표현하시오 
###### 그래프 제목: "과목별 점수" 
###### x축: 과목 / y축: 점수 
###### 막대 위에 점수 값 표시
###### 아래 데이터를 이용해 작성하시오.
###### python 항목 부분에만 레이블 값이 출력되도록 수정

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    '과목' :  ["Java", "Python", "React", "SQL"],
    "점수": [85, 90, 75, 80]
})

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(7,5))

bars = plt.bar(df['과목'], df['점수'], width = 0.5)

# for i in range(len())
for i in range(len(bars)):
    bar = bars[i]
    height = bar.get_height()

    if df['과목'][i] == 'Python':
        plt.text(bar.get_x() + bar.get_width() /2, height, f'{height}', ha ='center' , va='bottom')    

plt.title('과목별 점수')
plt.xlabel('과목')
plt.ylabel('점수')

plt.show()
```

###### 공부시간을 가로 막대그래프로 표현하시오
###### 막대 두께를 0.5로 설정
###### 그래프 제목: "학생별 공부시간“
###### 아래 데이터를 이용해 작성하시오.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "이름": ["김학생", "이학생", "박학생", "최학생"],
    "공부시간": [3, 5, 2, 4]
})

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(8,5))

df = df[df["이름"] != "김학생"]

plt.barh(df["이름"], df["공부시간"], height=0.5)

plt.title('학생별 공부시간')
plt.xlabel("공부시간")
plt.ylabel("이름")

plt.show()
```

######  매출 데이터를 선그래프로 표현

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "월": ["1월", "2월", "3월", "4월"],
    "매출": [100, 150, 130, 170]
})

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(5,3))

plt.plot(df["월"], df["매출"], marker = 'o')
plt.fill_between(df["월"], df["매출"],alpha = 0.3)
plt.title("월별 매출 변화")
plt.xlabel("월")
plt.ylabel("매출")

plt.show()
```

###### • 금액을 기준으로 원형 그래프 생성
###### • 퍼센트 표시
###### • 시작 각도 90도 설정
###### • 그래프 제목: "지출 비율“
###### • 아래 데이터를 이용해 작성하시오.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "카테고리": ["식비", "교통", "쇼핑", "기타"],
    "금액": [300000, 100000, 200000, 150000]
})

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(5,3))

plt.pie(
    df["금액"],
    labels = df["카테고리"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("지출 비율")

plt.show()
```

###### 6번
###### 두 데이터를 부서 기준으로 병합하시오(outer 조인) - 
###### 결측치는 각 컬럼의 평균값으로 처리하시오 - 
###### 기본급과 성과급 수당을 이용하여 누적 세로 막대그래프를 생성하시오 
###### 그래프 구성 : 아래 - 기본급 / 위 - 성과급 / 수당
###### 범례(legend 추가) (누적)
###### 그래프 제목 : 부서별 급여 구조
###### x축: 부서 / y축: 금액
###### 그래프를 이미지로 저장 (salary_chart.png, dpi=300)

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "인사팀"],
    "기본급": [300, 250, 200]
})

df2 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "마케팅팀"],
    "성과급": [100, 80, 70]
})

df3 = pd.DataFrame({
    "부서" : ["개발팀", "영업팀", "인사팀"],
    "수당" : [50, 40, 30]
})

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10,8))

# outer 조인
merged = pd.merge(df1, df2, on="부서",how= "outer")
merged02 = pd.merge(merged,df3,on="부서",how="outer")
merged02

# 결측치 제거
merged02["기본급"] = merged02["기본급"].fillna(merged02["기본급"].mean())
merged02["수당"] = merged02["수당"].fillna(merged02["수당"].mean())
merged02["성과급"] = merged02["성과급"].fillna(merged02["성과급"].mean())
merged02

plt.bar(merged02["부서"], merged02["기본급"], label="기본급")

plt.bar(
    merged02["부서"],
    merged02["성과급"],
    bottom=merged02["기본급"],
    label = "성과급"
)

plt.bar(
    merged02["부서"],
    merged02["수당"],
    bottom = (merged02["성과급"]+merged02["기본급"]),
    label="수당"
)

plt.legend()

plt.show()
```
