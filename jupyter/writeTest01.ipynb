score_data.csv파일의 데이터 중 국어 점수가 85점 이상인 데이터만 추출하여 출력
# DictReader => csv파일을 읽기 좋게 변환하는데 첫줄을 제목(필드명)으로 인식하게 변경해준다.

```python
import csv
# 1. 파일 읽기 
with open('greenData/score_data.csv','r',encoding='utf-8-sig') as f:
    # 2. csv를 읽기 좋은 방식인 DictReader를 사용한다.
    reader = csv.DictReader(f)
    print('국어 점수가 85점 이상인 데이터 : ')

    # 3. 한줄씩 확인하면서 if로 조건 검사
    for r in reader:
        # 파일에서 읽어온 모든 숫자는 문자열이 된다. => int() 를 통해 정수형으로 변환해준다.
        if int(r['국어']) >= 85:
            print(r)
```
