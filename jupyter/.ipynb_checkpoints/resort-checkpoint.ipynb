#### 회원 성별/나이별과 월별 가입자에 대한 그래프 
#### 전체회원, 남성, 여성
#### 월별 가입자수 - 표식 누적이 있는 꺾은선(범례:나이대,성별 - 범주:가입일 - 값 : 가입월 개) - member / join 필요없음

```python
import pandas as pd

df = pd.read_csv('greenData/resort/member.csv', encoding='utf-8-sig')
df2 = pd.DataFrame(df)
df2 = df2[df2["m_nickName"] != "admin"] # 관리자 제외

# m_regDate를 날짜형식으로 변경
df2['m_regDate'] = pd.to_datetime(df2['m_regDate'])

df2['가입연월'] = df2['m_regDate'].dt.strftime('%Y - %m')
df2.groupby('가입연월')['m_code'].count()
```

#### 예약이 많은 시기 - 막대그래프 - x축(월) -  y축(예약 수) - reservation / join 필요없음

```python
import pandas as pd

df = pd.read_csv('greenData/resort/reservation.csv', encoding='cp949')
df2 = pd.DataFrame(df)
df2 = df2[df2['cancel'] != 1]

# reserved_at를 날짜형식으로 변경
df2['reserved_at'] = pd.to_datetime(df2['reserved_at'])

df2['예약연월'] = df2['reserved_at'].dt.strftime('%Y - %m')
df2
df2.groupby('예약연월')['re_code'].count()
```

#### 할인중인 호텔 예약률, 비할인 호텔 예약률 → 원형 - reservation / join 필요없음

```python
import pandas as pd

re = pd.read_csv('greenData/resort/reservation.csv', encoding='cp949')
ho = pd.read_csv('greenData/resort/hotel.csv', encoding='utf-8-sig')
ro = pd.read_csv('greenData/resort/room.csv', encoding='cp949')

mergeReRo = pd.merge(re,ro,on='r_code',how='inner')
mergeReRoHo = pd.merge(ho,mergeReRo,on='h_code',how='inner')
mergeReRoHo

mergeReRoHo.groupby('city')["re_code"].count().head(10)
```

```python
import pandas as pd

re = pd.read_csv('greenData/resort/reservation.csv', encoding='cp949')
ro = pd.read_csv('greenData/resort/room.csv', encoding='cp949')

mergeReRo = pd.merge(re,ro,on='r_code',how='inner')
mergeReRo['minprice'] = mergeReRo.groupby("h_code")["price"].min()


for i in range(len(mergeReRo)):
    if 0 <= mergeReRo['minprice'] & mergeReRo['minprice'] < 100000:
        mergeReRo['가격대'] = '10만원'
    elif 100000 <= mergeReRo['minprice'] & mergeReRo['minprice'] < 200000:
        mergeReRo['가격대'] = '20만원'
    else:
        mergeReRo['가격대'] = '30만원'
    
```
