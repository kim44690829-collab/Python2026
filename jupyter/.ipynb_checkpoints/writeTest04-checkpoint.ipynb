pandas 라이브러리는 반드시 import해서 사용한다.
# CSV 파일은 공공데이터 포털에서 다운로드 받은 자료 => encoding='cp949'


```python
import pandas as pd

# 출력 옵션 설정
pd.set_option('display.max_columns',None) # 모든 컬럼 출력
pd.set_option('display.width',1000) # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width',True) # 한글 정렬 맞추기
# bike02.csv 를 pandas를 이용해서 읽기
df = pd.read_csv('greenData/bike02.csv', encoding = 'cp949')

# 상위 데이터 10개 출력
print("상위 10개 데이터")
print(df.head(10))
print()

# 대여 건수 1000 이상
print('대여건수 1000 이상')
print(df[df['대여건수'] >= 1000])
print()

# 대여건수 내림차순 정렬
print('대여건수 내림차순 정렬')
sort = df.sort_values(by='대여건수', ascending = False)
print(sort)
print()

# 대여건수 전체 합계 => 반드시 groupby해야함
total = df['대여건수'].sum()
avg = df['대여건수'].mean()
print('전체합계')
print(total)
print()

print('전체평균')
print(f"{avg:,.2f}")
print()

# 대여건수가 가장 많은 날짜 찾기
topdate = df.sort_values(by='대여건수', ascending= False)
print("대여건수가 가장 많은 날짜", topdate.iloc[0]['대여일자'])
```

```python
import pandas as pd

# 출력 옵션 설정
pd.set_option('display.max_columns',None) # 모든 컬럼 출력
pd.set_option('display.width',1000) # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width',True) # 한글 정렬 맞추기

print("--- 미세먼지 및 초미세먼지 주의지역 ---")
df = pd.read_csv('greenData/pm10.csv',encoding='cp949')
df01 = df[df['구분'] != '평균']
# 다중 조건을 줘야할때 and : & , or : **, not : ~
condition = (df01['미세먼지(PM10)'] >= 25) & (df01['초미세먼지(PM25)'] >= 15)
cond02 = df01[condition]
result = cond02[['구분','미세먼지(PM10)','초미세먼지(PM25)']]
print(result)
# df02 = df01[df01['미세먼지(PM10)'] >= 25]
# df03 = df02[df02['초미세먼지(PM25)'] >= 15]
# print(df03)
```

pandas에서 날짜를 추출하는 함수 종류
csv에 입력된 모든 날짜는 문자열 데이터 타입이다.
날짜형식으로 형변환하는 함수 : datetime()
# dt.month : 월 / dt.year : 년 / dt.day : 일 / dt.hour : 시 / dt.minute : 분 / dt.second : 초
# dt.dayofweek : 요일을 숫자로 반환 (0월 ~ 6일)
# dt.day_name() : 요일 이름이 문자열 ('monday')
# dt.quarter : 분기(1분기 ~ 4분기)

```python
import pandas as pd
# from datetime import datetime

pd.set_option('display.max_columns',None) # 모든 컬럼 출력
pd.set_option('display.width',1000) # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width',True) # 한글 정렬 맞추기\

df = pd.read_csv('greenData/accident.csv', encoding = 'cp949')

# 사고일자를 날짜형식으로 변경
df['사고일자'] = pd.to_datetime(df['사고일자'])

# 사고일자의 월을 출력
df['사고월'] = df['사고일자'].dt.month
# groupby().count() => 데이터가 비어있지않은 값의 개수를 센다.
# groupby().size() => 그룹별 전체 행의 개수를 센다.
# monthcount = df.groupby('사고월')['사고월'].sum().sort_values(ascending = False)
monthcount = df.groupby('사고월').size()
# idxmax => max값이 담긴 index를 가져옴
maxMonth = monthcount.idxmax()

# print(monthcount)
# print("사고가 가장 많이 난 월 : ", df01)
print("사고가 가장 많이 난 월 : ", maxMonth)

df['사상자수'] = df['사망'] + df['부상']
print(df)

result = df.groupby('노선명')['사상자수'].sum().sort_values(ascending = False).head(5)
print(result)

result02 = len(df[(df['원인'] == '졸음') & (df['사망'] >= 1)])
print(result02)
```
