```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------------
# 반드시 한극 깨짐 현상 코드 삽입
# -------------------------------------

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

data= pd.DataFrame({
    "수학" : [80,90,70],
    "영어" : [85,60,75],
    "국어" : [90,88,95]
}, index=["김학생", "이학생", "박학생"])

# heatmap()
# 1. 데이터 값을 색으로 표현하는 그래프
# 2. 숫자가 작을수록 => 진한색
# 3. 숫자가 클수록 => 연한색

sns.heatmap(data)

plt.title('학생 성적 히트맵', pad=20)

plt.show()
```
